```python
import torch
import os

# --- SỬA LỖI TRỌNG SỐ (PHIÊN BẢN ĐÃ SỬA) ---

def fix_state_dict_keys(state_dict):
    """Hàm này loại bỏ tiền tố 'stages.' khỏi các key."""
    new_state_dict = {}
    for k, v in state_dict.items():
        if k.startswith('stages.'):
            # Tạo key mới bằng cách bỏ đi 'stages.'
            new_key = k[len('stages.'):]
            new_state_dict[new_key] = v
        else:
            # Giữ nguyên các key khác
            new_state_dict[k] = v
    return new_state_dict

weights_path = '/kaggle/working/few-shot-scanpath/data/M2F_R50.pkl'
print(f"Đang tải trọng số từ: {weights_path}")

# Tải trọng số
original_weights = torch.load(weights_path)

# *** THAY ĐỔI 1: Xử lý trực tiếp `original_weights` thay vì `original_weights['model']` ***
fixed_weights = fix_state_dict_keys(original_weights)

# *** THAY ĐỔI 2: Lưu lại trực tiếp dictionary đã sửa, không cần tạo dict mới ***
torch.save(fixed_weights, weights_path)

print("Đã sửa và lưu lại file trọng số M2F_R50.pkl thành công!")
print("-" * 30)
```

```python
import os

# Đường dẫn tới file model cần sửa
model_file_path = '/kaggle/working/few-shot-scanpath/SE-Net/src/models.py'

# Đọc nội dung file
with open(model_file_path, 'r') as f:
    lines = f.readlines()

# Tìm và bình luận (comment out) các dòng code sửa đổi key không cần thiết
new_lines = []
for line in lines:
    # Các dòng này đang cố gắng đổi tên các key, gây ra lỗi. 
    # Vì chúng ta đã tự sửa file .pkl, chúng ta không cần chúng nữa.
    if 'bb_weights_new = bb_weights.copy()' in line:
        new_lines.append('# ' + line)
    elif 'for k, v in bb_weights.items():' in line:
        new_lines.append('# ' + line)
    elif "if k[:3] == 'res':" in line:
        new_lines.append('# ' + line)
    elif 'bb_weights_new["stages." + k] = v' in line:
        new_lines.append('# ' + line)
    elif 'bb_weights_new.pop(k)' in line:
        new_lines.append('# ' + line)
    # Dòng này cần được thay đổi để sử dụng trực tiếp trọng số đã sửa
    elif 'self.backbone.load_state_dict(bb_weights_new)' in line:
        # Thay thế bằng dòng load trực tiếp, bỏ qua việc tạo bb_weights_new
        new_lines.append('        self.backbone.load_state_dict(bb_weights)\n')
    else:
        new_lines.append(line)

# Ghi lại nội dung đã sửa
with open(model_file_path, 'w') as f:
    f.writelines(new_lines)

print("Đã sửa lỗi logic nạp trọng số trong src/models.py thành công!")
print("-" * 30)
```

```python
import os

# Đường dẫn đến file test.py cần ghi đè
file_path = '/kaggle/working/few-shot-scanpath/ISP/OSIE/GazeformerISP/src/test.py'

# Toàn bộ nội dung đúng của file test.py mà bạn cung cấp
correct_content = """
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import transforms
import numpy as np
import scipy.stats
import os
import argparse
import json
from tqdm import tqdm
from dataset.dataset import OSIE_evaluation
from utils.evaluation import comprehensive_evaluation_by_subject
from utils.logger import Logger
from utils.data_postprocess import *
from models.sampling import Sampling
from models.gazeformer import gazeformer
from models.models import Transformer

# ======================
# ARGUMENTS
parser = argparse.ArgumentParser(description="Scanpath prediction for OSIE dataset (CPU mode)")
parser.add_argument("--mode", type=str, default="test")
parser.add_argument("--img_dir", type=str, default="../../../data/OSIE", help="Path to image data")
parser.add_argument("--feat_dir", type=str, default="src/data/image_features", help="Path to image features")
parser.add_argument("--emb_dir", type=str, default="src/data/embeddings.npy", help="Path to embeddings file")
parser.add_argument("--fix_dir", type=str, default="src/data/fixations.json", help="Path to fixations file")
parser.add_argument("--evaluation_dir", type=str, default="src/assets/OSIE-ex-10to15", help="Evaluation assets path")
parser.add_argument("--user_emb_path", type=str, default="../../../data/assets/OSIE-ex-10to15/fewshot_user_embedding_1.pt")
parser.add_argument("--ex_subject", nargs='+', type=int, default=[-1], help='Subjects to exclude')
parser.add_argument("--fewshot_subject", nargs='+', type=int, default=[10, 11, 12, 13, 14], help='Subjects for few-shot evaluation')
# Model and evaluation config
parser.add_argument("--width", type=int, default=512)
parser.add_argument("--height", type=int, default=384)
parser.add_argument("--origin_width", type=int, default=800)
parser.add_argument("--origin_height", type=int, default=600)
parser.add_argument("--im_h", type=int, default=24)
parser.add_argument("--im_w", type=int, default=32)
parser.add_argument("--batch", type=int, default=1)
parser.add_argument("--seed", type=int, default=0)
parser.add_argument("--eval_repeat_num", type=int, default=1)
parser.add_argument("--min_length", type=int, default=1)
parser.add_argument("--max_length", type=int, default=8)
# Transformer params
parser.add_argument("--num_encoder", type=int, default=6)
parser.add_argument("--num_decoder", type=int, default=6)
parser.add_argument("--hidden_dim", type=int, default=512)
parser.add_argument("--nhead", type=int, default=8)
parser.add_argument("--img_hidden_dim", type=int, default=2048)
parser.add_argument("--lm_hidden_dim", type=int, default=768)
parser.add_argument("--encoder_dropout", type=float, default=0.1)
parser.add_argument("--decoder_dropout", type=float, default=0.1)
parser.add_argument("--cls_dropout", type=float, default=0.2)
parser.add_argument("--subject_feature_dim", type=int, default=384)
parser.add_argument("--action_map_num", type=int, default=4)
parser.add_argument("--subject_num", type=int, default=5)
parser.add_argument("--num_fewshot", type=int, default=1)
parser.add_argument("--random_support", type=int, default=0)
args = parser.parse_args([])
# FIX SEED + DEVICE
np.random.seed(args.seed)
torch.manual_seed(args.seed)
torch.backends.cudnn.benchmark = False
torch.backends.cudnn.deterministic = True
device = torch.device("cpu")
# TRANSFORM
transform = transforms.Compose([
    transforms.Resize((args.height // 2, args.width // 2)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
# MAIN FUNCTION
def main():
    # setup logging
    log_dir = args.evaluation_dir
    checkpoints_dir = os.path.join(log_dir, "checkpoints")
    os.makedirs(checkpoints_dir, exist_ok=True)
    log_info_folder = os.path.join('result', log_dir.split('/')[-1], "log")
    os.makedirs(log_info_folder, exist_ok=True)
    log_file = os.path.join(log_info_folder, "log_test_CPU.txt")
    logger = Logger(log_file)
    logger.info("Running OSIE evaluation on CPU")
    # ======================
    # LOAD DATA
    test_dataset = OSIE_evaluation(
        args, args.img_dir, args.feat_dir, args.fix_dir, args.emb_dir,
        action_map=(args.im_h, args.im_w),
        resize=(args.height, args.width),
        type="test", transform=transform
    )
    test_loader = DataLoader(
        dataset=test_dataset,
        batch_size=args.batch,
        shuffle=False,
        num_workers=0,
        collate_fn=test_dataset.collate_func
    )
    # LOAD MODEL
    transformer = Transformer(
        num_encoder_layers=args.num_encoder,
        nhead=args.nhead,
        subject_feature_dim=args.subject_feature_dim,
        d_model=args.hidden_dim,
        num_decoder_layers=args.num_decoder,
        encoder_dropout=args.encoder_dropout,
        decoder_dropout=args.decoder_dropout,
        dim_feedforward=args.hidden_dim,
        img_hidden_dim=args.img_hidden_dim,
        lm_dmodel=args.lm_hidden_dim,
        device=device,
        args=args
    ).to(device)
    model = gazeformer(
        transformer,
        spatial_dim=(args.im_h, args.im_w),
        args=args,
        subject_num=args.subject_num,
        action_map_num=args.action_map_num,
        dropout=args.cls_dropout,
        max_len=args.max_length
    ).to(device)
    sampling = Sampling(
        convLSTM_length=args.max_length,
        min_length=args.min_length,
        map_width=args.im_w,
        map_height=args.im_h,
        width=args.width,
        height=args.height
    )
    # LOAD CHECKPOINT
    ckpt_path = os.path.join(checkpoints_dir, "checkpoint_best.pth")
    if os.path.exists(ckpt_path):
        test_checkpoint = torch.load(ckpt_path, map_location=device)
        model_state = test_checkpoint.get('model', test_checkpoint)
        model.load_state_dict(model_state, strict=False)
        logger.info(f"Loaded checkpoint from {ckpt_path}")
    else:
        logger.info("WARNING: No checkpoint found, running with random weights!")
    model.eval()
    # EVALUATION LOOP
    all_gt_fix_vectors, all_predict_fix_vectors = [], []
    with tqdm(total=len(test_loader)) as pbar_test:
        for i_batch, batch in enumerate(test_loader):
            tmp = [batch["images"], batch["fix_vectors"], batch["task_embeddings"], batch["subjects"]]
            tmp = [_ if not torch.is_tensor(_) else _.to(device) for _ in tmp]
            tmp = [_.view(-1, *_.shape[2:]) if torch.is_tensor(_) else _ for _ in tmp]
            images, gt_fix_vectors, task_embeddings, subjects = tmp
            with torch.no_grad():
                predict = model(src=images, subjects=subjects, task=task_embeddings)
            log_normal_mu = predict["log_normal_mu"]
            log_normal_sigma2 = predict["log_normal_sigma2"]
            all_actions_prob = predict["all_actions_prob"]
            image_prediction_dict = {_: [] for _ in range(len(batch["img_names"]))}
            all_gt_fix_vectors.extend(gt_fix_vectors)
            for trial in range(args.eval_repeat_num):
                samples = sampling.random_sample(all_actions_prob, log_normal_mu, log_normal_sigma2)
                prob_sample_actions = samples["selected_actions_probs"]
                durations = samples["durations"]
                sample_actions = samples["selected_actions"]
                sampling_random_predict_fix_vectors, _, _ = sampling.generate_scanpath(
                    images, prob_sample_actions, durations, sample_actions)
                for idx in range(len(batch["img_names"])):
                    image_prediction_dict[idx].extend(
                        sampling_random_predict_fix_vectors[idx * args.subject_num:(idx + 1) * args.subject_num])
            all_predict_fix_vectors.extend(list(image_prediction_dict.values()))
            pbar_test.update(1)
            
    # COMPUTE METRICS
    cur_metrics, _, _ = comprehensive_evaluation_by_subject(
        all_gt_fix_vectors, all_predict_fix_vectors, args)
    
    logger.info("Evaluation metrics:")
    for metrics_key in cur_metrics.keys():
        for (metric_name, metric_value) in cur_metrics[metrics_key].items():
            logger.info(f"{metrics_key}-{metric_name}: {metric_value:.4f}")

    SM = scipy.stats.hmean(list(cur_metrics["ScanMatch"].values()))
    MM = np.mean(list(cur_metrics["MultiMatch"].values()))
    SED = cur_metrics["VAME"]["SED"]
    print(f'✅ SM: {round(SM, 3)}, MM: {round(MM, 3)}, SED: {round(SED, 3)}')

if __name__ == "__main__":
    main()
"""

# Ghi đè file test.py với nội dung đã được sửa
with open(file_path, 'w') as f:
    f.write(correct_content)
    
print(f"Đã ghi đè và sửa lỗi file {file_path} thành công!")
```
